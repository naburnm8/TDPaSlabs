Пример использования линейной регрессии

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Данные: Площадь (кв.м) vs. Цена (тыс. долларов)
X = np.array([50, 60, 70, 80, 90, 89, 121, 130]).reshape(-1, 1)
y = np.array([150, 180, 210, 240, 270, 300, 340, 400])

# Создание модели линейной регрессии
model = LinearRegression()
model.fit(X, y)

# Предсказание
X_test = np.array([100]).reshape(-1, 1)
y_pred = model.predict(X_test)
print(f"Предсказанная цена для 100 кв.м: {y_pred[0]} тыс.")

# Визуализация
plt.scatter(X, y, color='blue')
plt.plot(X, model.predict(X), color='red')
plt.xlabel('Площадь (кв.м)')
plt.ylabel('Цена (тыс. долларов)')
plt.title('Линейная регрессия')
plt.show()
```

Пример использования полиномиальной регрессии

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# Даные: Площадь (кв.м) vs. Цена (тыс. долларов)
X = np.array([50, 60, 70, 80, 90, 89, 125, 130]).reshape(-1, 1)
y = np.array([150, 120, 200, 240, 270, 300, 310, 400])

# Полиномиальные признаки степени 2
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Создание модели полиномиальной регрессии
model = LinearRegression()
model.fit(X_poly, y)

# Предсказание для 100 кв.м
X_test = np.array([100]).reshape(-1, 1)
X_test_poly = poly.transform(X_test)
y_pred = model.predict(X_test_poly)
print(f"Предсказанная цена для 100 кв.м: {y_pred[0]} тыс.")

# Визуализация
plt.scatter(X, y, color='blue')
plt.plot(X, model.predict(X_poly), color='red')
plt.xlabel('Площадь (кв.м)')
plt.ylabel('Цена (тыс. долларов)')
plt.title('Полиномиальная регрессия')
plt.show()
```
